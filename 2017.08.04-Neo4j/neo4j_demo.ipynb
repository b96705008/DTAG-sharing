# 2. Create Neo4j Data

```python
from py2neo import Graph, Node, Relationship
from py2neo.ogm import GraphObject, Property, RelatedTo, RelatedFrom
```

```python
class Customer(GraphObject):
    __primarykey__ = "customer_id"
    
    customer_id = Property()
    cust_json_prop = Property()
    
#     cust_type = Property()
#     age = Property()
#     gender_type_code = Property()
#     contact_zip_code = Property()
#     education_code = Property()
#     marital_status_code = Property()
#     salary_dp_cust_flag = Property()
#     wm_ao_definite_ind = Property()
#     employee_ind = Property()
#     y1_use_withhold_flag = Property()
#     sec_dp_cust_ind = Property()
#     birthday = Property()
#     start_date = Property()
#     start_year = Property()
#     register_zip_code = Property()
#     occupation = Property()
#     customer_group_code = Property()
#     customer_class_code = Property()
    
    road_help = RelatedTo("Customer")
    guarantee = RelatedTo("Customer")
    insure = RelatedTo("Customer")
    credit_card = RelatedTo("Customer")
    transfer = RelatedTo("Customer")
```

```python
class CathaySNA(object):
    def __init__(self):
        '''
        the Graph class which represents a Neo4j graph database instance
        '''
        self.graph = Graph(host="localhost", http_port=7474, password="801011")
        
        
    def new_customer(self, customer_id, property_json):  # property_dict
        customer = Customer()
        customer.customer_id = customer_id
        customer.cust_json_prop = property_json
        
#         customer.customer_id = property_dict['customer_id']
#         customer.cust_type = property_dict['cust_type']
#         customer.age = property_dict['age']
#         customer.gender_type_code = property_dict['gender_type_code']
#         customer.contact_zip_code = property_dict['contact_zip_code']
#         customer.education_code = property_dict['education_code']
#         customer.marital_status_code = property_dict['marital_status_code']
#         customer.salary_dp_cust_flag = property_dict['salary_dp_cust_flag']
#         customer.wm_ao_definite_ind = property_dict['wm_ao_definite_ind']
#         customer.employee_ind = property_dict['employee_ind']
#         customer.y1_use_withhold_flag = property_dict['y1_use_withhold_flag']
#         customer.sec_dp_cust_ind = property_dict['sec_dp_cust_ind']
#         customer.birthday = property_dict['birthday']
#         customer.start_date = property_dict['start_date']
#         customer.start_year = property_dict['start_year']
#         customer.register_zip_code = property_dict['register_zip_code']
#         customer.occupation = property_dict['occupation']
#         customer.customer_group_code = property_dict['customer_group_code']
#         customer.customer_class_code = property_dict['customer_class_code']
        
        self.graph.push(customer)  # Push data from one or more entities to their remote counterparts.

        
    def find_customer_by_id(self, customer_id):
        return Customer.select(self.graph).where("_.customer_id = '{}'".format(customer_id)).first()
    
    
    def update_relation(self, pri_customer_id, rel_customer_id, relation_type, relation_property_dict):
        pri_customer = self.find_customer_by_id(pri_customer_id)
        rel_customer = self.find_customer_by_id(rel_customer_id)
            
        if relation_type == '車籍資料相同':
            pri_customer.road_help.add(rel_customer, relation_property_dict)        
        elif relation_type == '放款保證人':
            pri_customer.guarantee.add(rel_customer, relation_property_dict)
        elif relation_type == '要被保人':
            pri_customer.insure.add(rel_customer, relation_property_dict)
        elif relation_type == '信用卡正附卡':
            pri_customer.credit_card.add(rel_customer, relation_property_dict)  
        elif relation_type == '存款轉帳':
            pri_customer.transfer.add(rel_customer, relation_property_dict)
        else:
            print "沒有該關係型態存在"
        
        self.graph.push(pri_customer)  # Push data from one or more entities to their remote counterparts.
    
    def clear_graph(self):
        '''
        清空neo4j db內容
        '''
        self.graph.delete_all()  # This method will permanently remove all nodes and relationships from the graph and cannot be undone.
```

```python
# 新增客戶節點資料
stime = time()

for index, row in cust_data.iterrows():
    property_dict = dict(row)
    property_json = json.dumps(property_dict, ensure_ascii=False, encoding='utf8')
    
    # cathay.new_customer(property_dict)
    cathay.new_customer(row['customer_id'], property_json)
    
print time() - stime
```

```python
# 新增客戶關係資料
stime = time()

for index, row in sna_data.iterrows():
    pri_customer_id = row['pri_customer_id']
    rel_customer_id = row['rel_customer_id']
    relation_type = row['SEG']
    relation_property_dict = {'level_cnt': row['level_cnt'], 'level_name': row['level_name']}

    cathay.update_relation(pri_customer_id, rel_customer_id, relation_type, relation_property_dict)
    
print time() - stime
```

# 3. Neo4j Cypher

```python
# 設定 degree: 不區分 in_degree, out_degree
cathay.graph.run("""
                  match (n:Customer)-[r]-()
                  with n, count(distinct r) as degree set n.deg = degree
                  return n.customer_id, n.deg
                  """).dump()
```

```python
# data(statement, parameters=None, **kwparameters)
# Run a Transaction.run() operation within an autocommit Transaction and extract the data as a list of dictionaries.
# The extracted data can then be easily passed into an external data handler such as a pandas.DataFrame for subsequent processing
result = cathay.graph.data("""
                            MATCH (n:Customer)
                            where n.deg >= 10
                            RETURN n
                            """)
```

```python
# 取出 degree >= 10 的客戶
target_ids = cathay.graph.data("""
                                MATCH (n:Customer)
                                where n.deg >= 10
                                OPTIONAL MATCH (n)-[r]-(m:Customer) 
                                RETURN {parent : n.customer_id, child : {customer_id :collect( m.customer_id)}}
                                """)
```
